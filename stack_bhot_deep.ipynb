```python
def mask_s2_clouds(image):
  qa = image.select('QA60')

  # Bits 10 and 11 are clouds and cirrus, respectively.
  cloud_bit_mask = 1 << 10
  cirrus_bit_mask = 1 << 11

  # Both flags should be set to zero, indicating clear conditions.
  mask = (
      qa.bitwiseAnd(cloud_bit_mask)
      .eq(0)
      .And(qa.bitwiseAnd(cirrus_bit_mask).eq(0))
  )

  return image.updateMask(mask).divide(10000)


dataset = (
    ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
    .filterDate('2020-01-01', '2020-01-30')
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
    .map(mask_s2_clouds)
)

visualization = {
    'min': 0.0,
    'max': 0.3,
    'bands': ['B4', 'B3', 'B2'],
}

m = geemap.Map()
m.set_center(83.277, 31.453, 12)
m.add_layer(dataset.mean(), visualization, 'RGB')
m
```

```python
import matplotlib.pyplot as plt
import numpy as np

def ndvi_to_color(ndvi_value):
  normalized_ndvi = (ndvi_value - (-1000.0)) / (5000.0 - (-1000.0))

  if normalized_ndvi >= 0.875:  # 012e01
    color = (1, 46, 1)
  elif normalized_ndvi >= 0.75:  # 207401
    color = (32, 112, 1)
  elif normalized_ndvi >= 0.625:  # 99b718
    color = (153, 183, 24)
  elif normalized_ndvi >= 0.5:  # 22cc04
    color = (34, 204, 4)
  elif normalized_ndvi >= 0.375:  # c6ca02
    color = (198, 202, 2)
  elif normalized_ndvi >= 0.25:  # fcd163
    color = (252, 208, 99)
  elif normalized_ndvi >= 0.125:  # ce7e45
    color = (206, 126, 69)
  else:  # White
    color = (255, 255, 255)

  return color


ndvi_values = np.linspace(-1000, 5000, 20)
colors = [ndvi_to_color(ndvi) for ndvi in ndvi_values]


plt.figure(figsize=(8, 2))
for i, ndvi in enumerate(ndvi_values):
  plt.bar(i, 1, color=[c / 255 for c in colors[i]])
  # plt.text(i, 0.5, f"{ndvi}", ha='center', va='center')

plt.xticks([])
plt.yticks([])
plt.title('NDVI to Color Mapping')
plt.show()
```

```python
import geemap
import ee
ee.Authenticate()
ee.Initialize()
dataset = ee.ImageCollection('NOAA/CDR/VIIRS/NDVI/V1') \
    .filterDate('2018-05-01', '2018-06-01')

ndvi = dataset.select('NDVI')
ndviVis = {
    'min': -1000.0,
    'max': 5000.0,
    'palette': [
        'ffffff', 'ce7e45', 'fcd163', 'c6ca02', '22cc04', '99b718', '207401',
        '012e01'
    ],
}
latitude = 7.71
longitude = 17.93
point = ee.Geometry.Point([longitude, latitude])
sample = ndvi.mean().sample(point, scale=1000).first()
ndvi_value = sample.get('NDVI').getInfo()
if ndvi_value is not None:
    print(f"NDVI value at ({latitude}, {longitude}): {ndvi_value}")
else:
    print(f"NDVI value not found at ({latitude}, {longitude})")

map_center = ee.Geometry.Point(longitude, latitude)
map_zoom = 6
m = geemap.Map(center=[map_center.getInfo()['coordinates'][1], map_center.getInfo()['coordinates'][0]], zoom=map_zoom)
m.addLayer(ndvi.mean(), ndviVis, 'NDVI')
```
